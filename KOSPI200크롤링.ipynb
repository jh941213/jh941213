```python
#kospi 200 차트
#일별시세 첫페이지 가져오기

import requests 
from bs4 import BeautifulSoup
import pandas as pd

#일별시세 목록
date, price, befores, per, volume, vprice = [], [], [], [], [], []
finance={'날짜':date,
         '가격':price,
         '전일비':befores,
         '등락률':per,
         '거래량':volume,
         '거래대금':vprice}

url = 'https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page=1'

resp = requests.get(url)
resp

#select

soup=BeautifulSoup(resp.content,'lxml')
date_tags=soup.select('td.date')
price_tags=soup.select('td.number_1')[::4]
before_tags=soup.select('td.number_1')[1::4]
per_tags=soup.select('td span.tah.p11')
volume_tags=soup.select('td.number_1')[2::4]
vprice_tags=soup.select('td.number_1')[3::4]

len(before_tags),len(per_tags),len(volume_tags),len(vprice_tags)

#데이터확인

for i in range(len(date_tags)):
    date.append(date_tags[i].text)
    price.append(price_tags[i].text)
    before = before_tags[i].text.strip()
    if before_tags[i]['class'][-1] == 'nv01':
        before=f'-{before}'
    befores.append(before)
    per.append(per_tags[i].text.strip())
    volume.append(volume_tags[i].text)
    vprice.append(vprice_tags[i].text)

finance
#데이터프레임화
kospi200 = pd.DataFrame(finance)

kospi200
```
