```python
import requests
from bs4 import BeautifulSoup
#덕배란 뉴스 기사 타이틀 가져오기
#1단계
#url 변수에 url 주소 넣기
url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%8D%95%EB%B0%B0"
#resp 에 get 시켜주기
resp = requests.get(url) 
resp #실행해서 200 이 나오면 정상적으로 가져옴
#2단계
```

```python
soup = BeautifulSoup(resp.content,'lxml')
```

```python
tags = soup.select('a.news_tit')
```

```python
tags

news_info={
    'title':[],
    'url':[]
}

for a_tag in tags:


    news_info['title'].append(a_tag['title'])
    news_info['url'].append(a_tag['href'])
```

```python
import pandas as pd # 저장하는 방법

df_news = pd.DataFrame(news_info)
df_news

df_news.to_csv('덕배뉴스정보.csv',encoding='utf-8-sig', index=False)
```

기사 제목 100개 가져오기
- requests.get(url) 10번
- 총 10페이지를 가져와야한다.
- 1page 에 1url

```python
import requests
from bs4 import BeautifulSoup
news_info={
    'title':[],
    'url' :[]
}

for i in range(10):
    
    start= str(i*10+1) # 변수기때문에 스트링으로 변환해줘서 뒤에 11,21값을 입력해줌
    url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%8D%95%EB%B0%B0&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=61&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start="+start
    resp = requests.get(url)
    
    soup = BeautifulSoup(resp.content,'lxml')
    a_tags = soup.select('a.news_tit')

    for a_tag in a_tags:
        news_info['title'].append(a_tag['title'])
        news_info['url'].append(a_tag['href'])


```
