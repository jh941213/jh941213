```python
#웹툰 요일, 제목, 평점, 작가,장르까지 포함한 데이터 프레임을 만들어라
import requests
from bs4 import BeautifulSoup
#url 1단계
url = 'https://comic.naver.com/webtoon/weekdayList?week=mon'

webtoon = {
    'title':[],
    'author':[],
    'rating':[],
    'day': []
}
weeks = ['mon','tue', 'wed', 'thu', 'fri', 'sat', 'dailyplus']
for week in weeks:
    url = 'https://comic.naver.com/webtoon/weekdayList?week='+week
    resp=requests.get(url)
    resp
    soup =BeautifulSoup(resp.content,'lxml')


    tit_tags=soup.select('ul.img_list li dl dt a')
    aut_tags=soup.select('ul.img_list li dl dd.desc')
    rat_tags=soup.select('ul.img_list li dl dd strong')

    for i in range(len(tit_tags)):
        webtoon['title'].append(tit_tags[i]['title'])
        webtoon['author'].append(aut_tags[i].text)
        webtoon['rating'].append(rat_tags[i].text)


webtoon

```

```python
#장르 
g_urls, g_names = [], []
url = 'https://comic.naver.com/webtoon/genre'
resp = requests.get(url)
soup = BeautifulSoup(resp.content,'lxml')
a_tags = soup.select('ul.spot a')

a_tags
for a_tag in a_tags:
    g_urls.append('https://comic.naver.com'+a_tag['href'])
#href(/webtoon/genre?genre=episode)로 코드가 되어 있기때문에 위에 구문
    g_names.append(a_tag.text)
g_urls, g_names
```

```python
from tqdm.notebook import tqdm
import pandas as pd
titles, authors, ratings, genres = [], [], [], []

for i in tqdm(range(len(g_urls))):
    genre, url = g_names[i], g_urls[i]
    resp = requests.get(url)
    # print(resp)

    soup = BeautifulSoup(resp.content, 'lxml')
    tit_tags = soup.select('ul.img_list li dl dt a')
    aut_tags = soup.select('ul.img_list li dl dd.desc a')
    rat_tags = soup.select('div.rating_type strong')

    for tit_tag, aut_tag, rat_tag in zip(tit_tags, aut_tags, rat_tags):
        titles.append(tit_tag['title'])
        authors.append(aut_tag.text)
        ratings.append(rat_tag.text)
        genres.append(genre)

dic_toons = {
    'title' : titles,
    'author' : authors,
    'rating' : ratings,
    'genre' : genres
}

df_toons = pd.DataFrame(dic_toons)
df_toons
```
